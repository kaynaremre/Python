```python
##Salih Emre Kaynar S011496
##This program has 3 functions in diffrent ranges which already defined. The purpose of this program is finding their 
##integral in two diffrent ways and compare with the actual result and plot the functions
##These two ways are numerical integration (midpoint rule) and integrate function from SciPy library
import numpy as np
import scipy.integrate
import matplotlib.pyplot as plt
def midpoint(f,a,b,n):  #The midpoint rule
    x=0
    y=0
    h=b-a
    h=h/n
    for i in range(0,n):
        x=f(a+i*h+1/2*h)
        y=y+x
    return h*y
def error(val,act): #The function will calculate the error percentages and absolute error values
    if act!=0:
        absErr=abs(val-act)
        errPcent=(absErr/act)*100
    else:
        absErr=abs(val-act)
        errPcent=0
    return absErr,errPcent
def plot (func,ranges,colour): #Plotting algorithm
    xs=range(0,1001)
    xss=[]
    for i in xs: #This loop will give a lot of x values to have a lot of solutions, thus we will have a smooth line
        xss.append(i/1000*ranges)
    ys=[]
    for j in xss: #This loop finds the y values from the domain x 
        ys.append(func(j))
    return plt.plot(xss, ys, color=colour)
## In the briefing it's saying that log(3) but a took it as ln(3) for easy calculation
print("f_1(x)=exp(x) integration range [0,log(3)]")
f1mid=midpoint(np.exp,0,np.log(3),100)
f1scipy=scipy.integrate.quad(np.exp,0,np.log(3))[0]
f1act=2
print("The result with midpoint is :", f1mid)
print("The result with scipy integrator is :", f1scipy)
print("Actual value is : ",f1act)
x=error(f1mid,f1act)
print("The error rate of midpoint value is :",x[0],"%   Absolute error is :",x[1])
x=error(f1scipy,f1act)
print("The error rate of SciPy value is :",x[0],"%   Absolute error is :",x[1])
plot(np.exp,np.log(3),'blue')
leg=['e^x in range 0 to ln3']
plt.legend(leg)
plt.show()
print("f_2(x)=cos(x) integration range [0,pi]")
f2mid=midpoint(np.cos,0,np.pi,100)
f2scipy=scipy.integrate.quad(np.cos,0,np.pi)[0]
f2act=0
print("The result with midpoint is :", f2mid)
print("The result with scipy integrator is :", f2scipy)
print("Actual value is : ",f2act)
x=error(f2mid,f2act)
print("The error rate of midpoint value is : ",x[1],"%   Absolute error is :",x[0])
x=error(f2scipy,f2act)
print("The error rate of SciPy value is : ",x[1],"%   Absolute error is :",x[0])
plot(np.cos,np.pi,'green')
leg=['cos(x) in range 0 to  ']
plt.legend(leg)
plt.show()
print("f_3(x)=sin(x) integration range [0,pi]")
f3mid=midpoint(np.sin,0,np.pi,100)
f3scipy=scipy.integrate.quad(np.sin,0,np.pi)[0]
f3act=2
print("The result with midpoint is :", f3mid)
print("The result with scipy integrator is :", f3scipy)
print("Actual value is : ",f3act)
x=error(f3mid,f3act)
print("The error rate of midpoint value is : ",x[1],"%   Absolute error is :",x[0])
x=error(f3scipy,f3act)
print("The error rate of SciPy value is : ",x[1],"%   Absolute error is :",x[0])
plot(np.sin,np.pi,'black')
leg=['sin(x) in range 0 to  ']
plt.legend(leg)
plt.show()
print("f_4(x)=sin(x) integration range [0,pi/2]")
print(midpoint(np.sin,0,(np.pi)/2,100))
f4mid=midpoint(np.sin,0,(np.pi)/2,100)
f4scipy=scipy.integrate.quad(np.sin,0,np.pi/2)[0]
f4act=1
print("The result with midpoint is :", f4mid)
print("The result with scipy integrator is :", f4scipy)
print("Actual value is : ",f4act)
x=error(f4mid,f4act)
print("The error rate of midpoint value is : ",x[1],"%   Absolute error is :",x[0])
x=error(f4scipy,f4act)
print("The error rate of SciPy value is : ",x[1],"%   Absolute error is :",x[0])
plot(np.sin,np.pi/2,'orange')
leg=['sin(x) in range 0 to  /2']
plt.legend(leg)
plt.show()
```
